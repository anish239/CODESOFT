```python
from datetime import datetime

train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])

train_df['trans_hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['trans_dayofweek'] = train_df['trans_date_trans_time'].dt.dayofweek 
train_df['trans_month'] = train_df['trans_date_trans_time'].dt.month

test_df['trans_hour'] = test_df['trans_date_trans_time'].dt.hour
test_df['trans_dayofweek'] = test_df['trans_date_trans_time'].dt.dayofweek
test_df['trans_month'] = test_df['trans_date_trans_time'].dt.month

train_df['dob'] = pd.to_datetime(train_df['dob'])
test_df['dob'] = pd.to_datetime(test_df['dob'])

train_df['age'] = (train_df['trans_date_trans_time'] - train_df['dob']).dt.days // 365
test_df['age'] = (test_df['trans_date_trans_time'] - test_df['dob']).dt.days // 365

train_df = train_df.drop(columns=['trans_date_trans_time', 'dob'])
test_df = test_df.drop(columns=['trans_date_trans_time', 'dob'])
print("Processed date/time columns and created hour, dayofweek, month, age features.")
```

```python
categorical_cols = ['merchant', 'category', 'gender', 'city', 'state', 'job']
numerical_cols = ['amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'trans_hour', 'trans_dayofweek', 'trans_month', 'age']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

categorical_to_encode = ['gender', 'category', 'state']
high_cardinality_to_drop = ['merchant', 'city', 'job']
numerical_cols = ['amt', 'lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'trans_hour', 'trans_dayofweek', 'trans_month', 'age'] # Ensure these match columns in your df
target_col = 'is_fraud'

print(f"Dropping high cardinality columns: {high_cardinality_to_drop}")
train_df = train_df.drop(columns=high_cardinality_to_drop)
test_df = test_df.drop(columns=high_cardinality_to_drop)
print("\nRemaining columns in train_df:", train_df.columns.tolist())

X_train = train_df.drop(columns=[target_col])
y_train = train_df[target_col]

X_test = test_df.drop(columns=[target_col])
y_test = test_df[target_col]
print(f"\nX_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_to_encode) 
    ],
    remainder='passthrough' 
)

print("\nFitting preprocessor on X_train...")
preprocessor.fit(X_train)

print("Transforming X_train...")
X_train_processed = preprocessor.transform(X_train)
print("Transforming X_test...")
X_test_processed = preprocessor.transform(X_test)

num_feature_names = numerical_cols
cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_to_encode).tolist()
processed_feature_names = num_feature_names + cat_feature_names

X_train_processed_df = pd.DataFrame(X_train_processed, columns=processed_feature_names, index=X_train.index)
X_test_processed_df = pd.DataFrame(X_test_processed, columns=processed_feature_names, index=X_test.index)

print(f"\nShape of processed X_train: {X_train_processed_df.shape}")
print(f"Shape of processed X_test: {X_test_processed_df.shape}")
print("\nFirst 5 rows of processed X_train:")
print(X_train_processed_df.head())

print("\nNaNs in processed X_train:", X_train_processed_df.isnull().sum().sum())
print("NaNs in processed X_test:", X_test_processed_df.isnull().sum().sum())
```

```python
from imblearn.over_sampling import SMOTE
import numpy as np 

print("Class distribution before SMOTE:")
print(y_train.value_counts())

print("\nApplying SMOTE to the training data...")
smote = SMOTE(random_state=42, n_jobs=-1)

X_train_resampled, y_train_resampled = smote.fit_resample(X_train_processed, y_train)
print("SMOTE finished.")

print("\nClass distribution after SMOTE:")
unique, counts = np.unique(y_train_resampled, return_counts=True)
print(dict(zip(unique, counts)))

print(f"\nShape of X_train_resampled: {X_train_resampled.shape}")
print(f"Shape of y_train_resampled: {y_train_resampled.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, auc, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns
import time 
log_reg = LogisticRegression(random_state=42, max_iter=1000, solver='liblinear')

dt_clf = DecisionTreeClassifier(random_state=42)

rf_clf = RandomForestClassifier(random_state=42, n_estimators=100, n_jobs=-1) # 100 trees is a common starting point

models = {
    "Logistic Regression": log_reg,
    "Decision Tree": dt_clf,
    "Random Forest": rf_clf
}

results = {}
for name, model in models.items():
    print(f"--- Training {name} ---")
    start_time = time.time()
    model.fit(X_train_resampled, y_train_resampled)

    end_time = time.time()
    training_time = end_time - start_time
    print(f"Training finished in {training_time:.2f} seconds.")

    print(f"\n--- Evaluating {name} on Test Set ---")
    y_pred = model.predict(X_test_processed)
    y_pred_proba = model.predict_proba(X_test_processed)[:, 1]
    print("\nClassification Report:")
    report = classification_report(y_test, y_pred, target_names=['Legitimate (0)', 'Fraud (1)'])
    print(report)

    print("Confusion Matrix:")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Legitimate', 'Fraud'], yticklabels=['Legitimate', 'Fraud'])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'{name} - Confusion Matrix')
    plt.show()
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    print(f"\nROC AUC Score: {roc_auc:.4f}")

    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    pr_auc = auc(recall, precision) 
    avg_precision = average_precision_score(y_test, y_pred_proba) 
    print(f"Precision-Recall AUC: {pr_auc:.4f}")
    print(f"Average Precision Score: {avg_precision:.4f}")

    results[name] = {
        "model": model,
        "y_pred": y_pred,
        "y_pred_proba": y_pred_proba,
        "classification_report": report,
        "confusion_matrix": cm,
        "roc_auc": roc_auc,
        "pr_auc": pr_auc,
        "avg_precision": avg_precision,
        "training_time": training_time
    }
    print("-"*(20 + len(name))) 

print("\n\n--- Model Comparison ---")
for name, metrics in results.items():
    print(f"\n{name}:")
    print(f"  Training Time: {metrics['training_time']:.2f}s")
    print(f"  ROC AUC: {metrics['roc_auc']:.4f}")
    print(f"  PR AUC: {metrics['pr_auc']:.4f}")
    print(f"  Average Precision: {metrics['avg_precision']:.4f}")

    print("  Key Metrics for Fraud Class (from Classification Report):")
    report_lines = metrics['classification_report'].split('\n')
    fraud_line = [line for line in report_lines if 'Fraud (1)' in line]
    if fraud_line:
        print(f"    {fraud_line[0].strip()}")
```
